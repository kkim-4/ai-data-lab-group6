```python
import pandas as pd

# Load the CSV file into a DataFrame
file_path = '/content/drive/MyDrive/EDA of the Chemistry Unbound Exit Report - Sheet2.csv'
df = pd.read_csv(file_path, index_col =0)

# Display the table
from IPython.display import display
display(df)
```

```python
# VADER setup

# Install and import nltk
!pip install nltk
import nltk

# Punkt tokenizer
nltk.download('punkt_tab')

# Download the lexicon
nltk.download("vader_lexicon")
```

```python
# Import the lexicon
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Create an instance of SentimentIntensityAnalyzer
sent_analyzer = SentimentIntensityAnalyzer()

# Example
sentence = "VADER is pretty good at identifying the underlying sentiment of a text!"
print(sent_analyzer.polarity_scores(sentence))
```

```python
def aggregate_sentiment_scores(text):
    """Calculates and returns the positive and negative sentiment scores for a given text.

    Args:
        text (str or list): The input text or a list of text chunks.

    Returns:
        tuple: A tuple containing the positive and negative sentiment scores.
    """
    try:
        if isinstance(text, list):
            text = ' '.join(str(chunk) for chunk in text)  # Join list elements safely into a string
        scores = sent_analyzer.polarity_scores(text)
        return scores['pos'], scores['neg']
    except (AttributeError, TypeError):
        return 0, 0  # Return 0 for both scores if an error occurs

# Apply the function to your text column(s)
for column in df.select_dtypes(include=['object']).columns:
    df[column + '_pos_score'] = df[column].apply(aggregate_sentiment_scores)
    df[column + '_pos_score'] = df[column + '_pos_score'].str[0].fillna(0)  # Handle missing values
    df[column + '_neg_score'] = df[column].apply(aggregate_sentiment_scores)
    df[column + '_neg_score'] = df[column + '_neg_score'].str[1].fillna(0)  # Handle missing values

# Display the DataFrame with aggregated scores
display(df)
```

```python
df.to_csv('/content/drive/My Drive/altered_df.csv', index=False)
```

```python
def format_output(output_dict):

  polarity = "neutral"

  if(output_dict['compound']>= 0.05):
    polarity = "positive"

  elif(output_dict['compound']<= -0.05):
    polarity = "negative"

  return polarity

def predict_sentiment(text):

  output_dict =  sent_analyzer.polarity_scores(text)
  return format_output(output_dict)
```

```python
def split_into_chunks(text, chunk_size=100):
    tokens = nltk.word_tokenize(text)
    chunks = [' '.join(tokens[i:i + chunk_size]) for i in range(0, len(tokens), chunk_size)]
    return chunks
```

```python
for column in df.select_dtypes(include=['object']).columns:
    df[column + '_chunks'] = df[column].apply(split_into_chunks)
    df[column + '_sentiment'] = df[column + '_chunks'].apply(lambda chunks: [predict_sentiment(chunk) for chunk in chunks])
```
